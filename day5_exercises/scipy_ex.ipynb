```python
import numpy as np
import scipy as sp

import matplotlib.pyplot as plt
```

```python
A = np.array([[1, -2, 3],
              [4, 5, 6],
              [7, 1, 9]])

b = np.array([1, 2, 3])

x = sp.linalg.solve(A, b)
print(x)
```

```python
poisson_dist = sp.stats.poisson(mu=3)
poisson_samples = poisson_dist.rvs(size=1000)

plt.hist(poisson_samples, density=True, label="samples")
x = np.arange(0, 10)
plt.plot(x, poisson_dist.pmf(x), 'ro-', label="PMF")
plt.plot(x, poisson_dist.cdf(x), 'bo-', label="CDF")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
normal_dist = sp.stats.norm(loc=4, scale=2.5)
normal_samples = normal_dist.rvs(size=1000)

plt.hist(normal_samples, density=True, bins=30, label="samples")
x = np.linspace(-5, 15, 100)
plt.plot(x, normal_dist.pdf(x), 'r-', label="PDF")
plt.plot(x, normal_dist.cdf(x), 'b-', label="CDF")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
sp.stats.ttest_ind(poisson_samples, normal_samples)
```
